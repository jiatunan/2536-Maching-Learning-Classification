## x1_x2
The E-step :
Responsibilities at iteration m for data point n in class y is:

$$
h_{n,y}^m = \frac{\pi_y^m \; f_{X_1,X_2|Y_n=y}(x_1,x_2; \mu_y^m, \sigma_y^{2(m)}, \lambda_y^m)}{\sum_{y\in{\mathcal Y}} \pi_y^m \;f_{X_1,X_2|Y_n=y}(x_1,x_2; \mu_y^m, \sigma_y^{2(m)},\lambda_y^m)}
$$

```python
# this function represents the e-step compute the expectation of complete loglikelihood
def Reponsibletitle_x1_x2(x1,x2,pi,mu,Sigma,lmbda):  #return the gamma for each (i,y)
    
    K = pi.size    #number of pi
    
    g = np.zeros((K, x1.size))
    
    for n in range(0,x1.size):         #loop for each xi
        
        denom = 0
                
        for k in range(0,K): #loop for each Y label
            
            g[k,n] = st.norm.pdf(x1[n], mu[k], math.sqrt(Sigma[k])) * pi[k]* lmbda[k]* np.exp(-x2[n]*lmbda[k])
            denom += g[k,n]
            
        for k in range(0,K):
            g[k,n] /= denom
            
   
    return g
```

```python
def EM_x1_x2(x1,x2, pi, mu, Sigma, lmbda, iter):   #for sigma, it refers to the variance of x1 here
    
    K = pi.size
    pi_hist=np.zeros((K,iter))
    mu_hist=np.zeros((K,iter))
    sigma_hist=np.zeros((K,iter))
    lmbda_hist=np.zeros((K,iter))
    for m in range(iter):
        
        # e-step
        gamma = Reponsibletitle_x1_x2(x1,x2,pi,mu,Sigma,lmbda) 
        
        numer = np.sum(gamma, axis=1) # sum_n  gamma_{n,k} gives array shape (1,K)
        denom = np.sum(numer) # sum_{n,k}  gamma_{n,k}
        pi = numer / denom   #update pi to the m+1 step
        
        numer_2 = np.matmul(gamma, x1)   # sum_n x_{n,l} gamma_{n,k} gives array shape (K,d) 
        sum_gamma = np.sum(gamma, axis=1) # sum_n  gamma_{n,k} gives array shape (1,K)
        
        for k in range(0,K):
            mu[k] = numer_2[k]/sum_gamma[k]  #update mu to the m+1 step
            
        for k in range(0,K):
            square=(x1-mu[k])**2
            numer_3=np.matmul(gamma, square)
            sum_gamma_2=np.sum(gamma,axis=1)
            Sigma[k]=numer_3[k]/sum_gamma_2[k]
        
        denom_2  = np.matmul(gamma, x2)   # sum_n x_{n,l} gamma_{n,k} gives array shape (K,d) 
        for k in range(0,K):
            lmbda[k] = numer[k]/denom_2[k]  #  compute update to lmbda
            
        pi_hist[:,m]=pi
        mu_hist[:,m]=mu
        sigma_hist[:,m]=Sigma
        lmbda_hist[:,m]=lmbda
            
    return  pi_hist,mu_hist,sigma_hist,lmbda_hist,pi,mu,Sigma,lmbda
```

## x1_x3
The E-step :
Responsibilities at iteration m for data point n in class y and X3 in category k is

$$
h_{n,y}^m = \frac{\pi_y^m \; f_{X_1,X_3|Y_n=y}(x_1,x_3; \mu_y^m, \sigma_y^{2(m)}, \gamma_{k,y}^m)}{\sum_{y\in{\mathcal Y}} \pi_y^m \;f_{X_1,X_3|Y_n=y}(x_1,x_3; \mu_y^m, \sigma_y^{2(m)},\gamma_{k,y}^m)}
$$

```python
# this function represents the e-step compute the expectation of complete loglikelihood
def Reponsibletitle_x1_x3(x1,x3,pi,mu,Sigma,cat):  #return the gamma for each (i,y)
    
    K = pi.size    #number of pi
    
    q= x3.unique().size
        
    g = np.zeros((K, x1.size))
        
    #ind=np.array(x3==q+1)    #creat a mask to filter all x3=k
    
    for n in range(0,x1.size):         #loop for each xi

        denom = 0

        for k in range(0,K): #loop for each Y label

            g[k,n] = st.norm.pdf(x1[n], mu[k], math.sqrt(Sigma[k])) *pi[k]*cat[x3[n]-1, k] 
                
            denom += g[k,n]

        for k in range(0,K):
            g[k,n] /= denom
                
        
    return g
```

```python
def EM_x1_x3(x1,x3, pi, mu, Sigma, cat, iter): 
    
    K = pi.size
    P=x3.unique().size
    
    pi_hist=np.zeros((K,iter))
    mu_hist=np.zeros((K,iter))
    sigma_hist=np.zeros((K,iter))
    cat_hist=np.zeros((P,K,iter))
    for m in range(iter):
        gamma = Reponsibletitle_x1_x3(x1,x3,pi,mu,Sigma,cat)    
        
        numer = np.sum(gamma, axis=1) # sum_n  gamma_{n,k} gives array shape (1,K)
        denom = np.sum(numer) # sum_{n,k}  gamma_{n,k}
        pi = numer / denom   #update pi to the m+1 step

        numer_2 = np.matmul(gamma, x1)   
        sum_gamma = np.sum(gamma, axis=1) 
        for k in range(0,K):
            mu[k] = numer_2[k]/sum_gamma[k]  #update mu to the m+1 step

        for k in range(0,K):
            square=(x1-mu[k])**2
            numer_3=np.matmul(gamma, square)
            sum_gamma_2=np.sum(gamma,axis=1)
            Sigma[k]=numer_3[k]/sum_gamma_2[k] #update the variance

        gamma=pd.DataFrame(gamma)
        for p in range(0, P):
            mask=gamma.iloc[:,list(x3==p+1)]
            numer_3=mask.sum(axis=1)
            denom=gamma.sum(axis=1)
                
            for k in range(0,K):
                cat[p,k]=numer_3[k]/denom[k] #update the cat
        pi_hist[:,m]=pi
        mu_hist[:,m]=mu
        sigma_hist[:,m]=Sigma
        cat_hist[:,:,m]=cat

    return  pi_hist, mu_hist, sigma_hist, cat_hist,pi,mu,Sigma,cat
```

## x2_x3
The E-step :
Responsibilities at iteration m for data point n in class y and X3 in category k is

$$
h_{n,y}^m = \frac{\pi_y^m \; f_{X_2,X_3|Y_n=y}(x_2,x_3; \lambda_y^m, \gamma_{k,y}^m)}{\sum_{y\in{\mathcal Y}} \pi_y^m \;f_{X_2,X_3|Y_n=y}(x_2,x_3; \lambda_y^m,\gamma_{k,y}^m)}
$$

```python
# this function represents the e-step compute the expectation of complete loglikelihood
def Reponsibletitle_x2_x3(x2,x3,pi,lmbda,cat):  #return the gamma for each (i,y)
    
    K = pi.size    #number of pi
    
    q= x3.unique().size
        
    g = np.zeros((K, x2.size))
        
    #ind=np.array(x3==q+1)    #creat a mask to filter all x3=k
    
    for n in range(0,x2.size):         #loop for each xi

        denom = 0

        for k in range(0,K): #loop for each Y label

            g[k,n] = pi[k]*cat[x3[n]-1, k] * lmbda[k]* np.exp(-x2[n]*lmbda[k]) 
                
            denom += g[k,n]

        for k in range(0,K):
            g[k,n] /= denom
                
        
    return g
```

```python
def EM_x2_x3(x2,x3,pi,lmbda, cat, iter):  
    
    K = pi.size
    P=x3.unique().size
    
    pi_hist=np.zeros((K,iter))
    lmbda_hist=np.zeros((K,iter))
    cat_hist=np.zeros((P,K,iter))
    
    
  
  
    for m in range(iter):
        
        # e-step
        gamma = Reponsibletitle_x2_x3(x2,x3,pi,lmbda,cat)    
        
        #m-step
        
        #  compute update to pi
        numer = np.sum(gamma, axis=1) 
        denom = np.sum(numer)
        pi = numer / denom   #update pi to the m+1 step
       

         #  compute update to lmbda
        denom_2  = np.matmul(gamma, x2)   
        for k in range(0,K):
            lmbda[k] = numer[k]/denom_2[k]
           

        #  compute update to cat
        gamma=pd.DataFrame(gamma)
        for p in range(0, P):
            mask=gamma.iloc[:,list(x3==p+1)]
            numer_3=mask.sum(axis=1)
            denom=gamma.sum(axis=1)
                
            for k in range(0,K):
                cat[p,k]=numer_3[k]/denom[k]
            
        pi_hist[:,m]=pi
        lmbda_hist[:,m]=lmbda
        cat_hist[:,:,m]=cat

            
    return pi_hist, lmbda_hist, cat_hist, pi,lmbda,cat
```

```python
def EM_x123(x, pi, mu, sigma, lmbda, cat, iter):  
    
    K = pi.size
    
    P=cat.shape[0]
 
    for m in range(iter):
        gamma = Responsibilitytitles_x123(x,pi,mu,sigma,lmbda,cat)   
        #m step
        numer = np.sum(gamma, axis=1) 
        denom = np.sum(numer) 
        pi = numer / denom   #update the prior function pi

        numer_2 = np.matmul(gamma, x1)  
        sum_gamma = np.sum(gamma, axis=1)
        for k in range(0,K):
            mu[k] = numer_2[k]/sum_gamma[k]  #update mu for x1
        for k in range(0,K):
            square=(x1-mu[k])**2
            numer_3=np.matmul(gamma, square)
            sum_gamma_2=np.sum(gamma,axis=1)
            sigma[k]=numer_3[k]/sum_gamma_2[k] #update the variance for x1

        denom_2  = np.matmul(gamma, x2)   
        for k in range(0,K):
            lmbda[k] = numer[k]/denom_2[k] #update the lambda for x2

        gamma=pd.DataFrame(gamma)
        for p in range(0, P):
            mask=gamma.iloc[:,list(x3==p+1)]
            numer_4=mask.sum(axis=1)
            denom=gamma.sum(axis=1)
                
            for k in range(0,K):
                cat[p,k]=numer_4[k]/denom[k]
        
        pi_hist[:,m]=pi
        mu_hist[:,m]=mu
        sigma_hist[:,m]=sigma
        lmbda_hist[:,m]=lmbda
        cat_hist[:,:,m]=cat

    return pi_hist, mu_hist, sigma_hist,lmbda_hist, cat_hist, pi, mu, sigma,lmbda, cat
```

#### Now we've set all EM algorithms, we're able to find the time when all parameters converge.

```python
def InitialGuess(x,K):
    # generate an initial guess depending on the number of classes that are specified

    pi_h = np.ones(K)/K
    mu_h = x.iloc[:,0].mean()
    Sigma_h = x.iloc[:,0].std()**2  #for x1 normal distribution
    
    lmbda_h=1/x.iloc[:,1].mean()  #for x2 exponential 
    
    C=x.iloc[:,2].unique().size
    cat_h=np.zeros(C)
    for c in range(0,C):
        cat_h[c]=np.sum(np.array(x.iloc[:,2]==c+1))/x.shape[0] #for x3 categorical 
        
    #now broadcast the singel estimate to 3 estimates

    pi0 = pi_h * np.random.uniform(0.99,1.01,K)
    pi0 = pi0 / sum(pi0)                       #normalize 

    mu0 = mu_h * np.random.uniform(0.99,1.01,K)
    Sigma0 = Sigma_h * np.ones(K)               #we don't uniform sigma here
    
    lmbda0=lmbda_h * np.random.uniform(0.99,1.01,K)
    
    cat0=np.zeros((C,K))
    for c in range(0,C):
        cat0[c,:]=cat_h[c]*np.random.uniform(0.99,1.01,K)
    for k in range(0,K):
        normalizer=np.sum(cat0[:,k])
        cat0[:,k] /= normalizer      #normalize 
    
    
    
    return pi0, mu0, Sigma0,lmbda0, cat0
```

```python
# compute the posterior probability of a point in feature space being from a given class
# generate an nxn grid, where (i,j) entry represents the probability of Y=y when 
# x1=i,x2=j

def Posterior_x12(pi,mu,Sigma,lmbda,bins):                #bins represents the rectangle containing the values of x1,x2
    K = pi.size
    
    p = np.zeros((K,bins.shape[1],bins.shape[1]))
    
    for i in range(0,bins.shape[1]):
        for j in range(0,bins.shape[1]):
                 
            denom = 0
    
            for k in range(0,K):
                 p[k,i,j] = st.norm.pdf(bins[0,i],mu[k],math.sqrt(Sigma[k]))*pi[k]*lmbda[k]*np.exp(-bins[1,j]*lmbda[k])
                 denom += p[k,i,j]

            for k in range(0,K):
                 p[k,i,j] /= denom
                 
    return p
```

## Using BIC Score to select the best model 

```python
def BIC(x,K,iter):
    
    pi,mu,sigma,lmbda,cat=InitialGuess(x,K) #initial guess
     
    
    pi_hist,,mu_hist,sigma_hist,lmbda_hist,cat_hist, pi,mu,sigma,lmbda,cat=EM_x123(x, pi,mu,sigma,lmbda,cat,iter) #ultimate parameters
    
    
    gamma=Responsibilitytitles_x123(x,pi,mu,sigma,lmbda,cat) #update the posterior
    
    
    x1=x.iloc[:,0]
    x2=x.iloc[:,1]
    x3=x.iloc[:,2]
    
    C=cat.shape[0]
    
    #loglikelihood for all 3 features
    
    ll=0
    for n in range(0,x.shape[0]):
        p=0
        for k in range(0,K):
            p+=pi[k]*st.norm.pdf(x1[n],mu[k],math.sqrt(sigma[k]))* lmbda[k]*np.exp(-x2[n]*lmbda[k])*cat[x3[n]-1,k]
        ll+=np.log(p)

                
    bic= -ll+0.5*((K*(3+(C-1)))+(K-1))*np.log(x.shape[0])
    return bic
    
    
```

```python
Kfit_list=np.arange(1,8,1)
bic_list=[]
for k in Kfit_list:
    bic_list.append(BIC(x,k,100))
    
#plot the result
plt.plot(Kfit_list,bic_list)
plt.title('BIC Score')
plt.ylabel(r'$BIC$')
plt.xlabel(r'$Regimes$')
```
